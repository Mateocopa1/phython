```python
import matplotlib.pyplot as plt
```

```python
# ejemplo de grafica con un solo eje
kilometros = [2.5, 3.1, 3.5, 3, 3.9, 4.8, 6.7, 9.4]

plt.plot(kilometros)
plt.show()
```

```python
# ejemplo con dos ejes
dias =['Lunes','Martes','Miercoles','Jueves','Viernes', 'Sabado','Domingo']
kilometros_2 = [2.5, 3.1, 3.5, 3.9, 4.8, 6.7, 9.4]

plt.plot(dias, kilometros_2, marker = "o", linestyle = '--', color ='black')
plt.title('Kilometros Recoridos')
plt.xlabel('Dias')
plt.ylabel('Kilometros')
plt.show()
```

```python
#Graficos de dispersion

kilometros_3 = [2.5, 3.1, 3.5, 3.9, 4.8, 6.7, 9.4]
calorias = [1500, 1300, 1200, 1000, 1100, 800, 750]

plt.scatter(calorias, kilometros_3, color='yellow')
plt.title('Kilometros VS Calorias')
plt.xlabel('Calorias')
plt.ylabel('Kilometros')
plt.show()
```

```python
#barras de error

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
errorx = [0.5, 0.3, 0.7, 0.2, 0.4]
errory = [0.1, 0.2, 0.7, 0.4, 0.6]

plt.errorbar(x,y, xerr=errorx, yerr=errory, fmt='o', color='red')
plt.title('Cant Ventas VS Millones Con Desviacion')
plt.xlabel('Cant Ventas')
plt.ylabel('Millones')
plt.show()
```

```python
# Ejemplo de Histograma
data = [2, 3, 5, 5, 5, 6, 6, 7, 8, 9, 10, 11, 11, 11, 12, 13, 15]

plt.hist(data, bins=7, width = 0.8, color = 'lightblue')
plt.show()
```

```python
import numpy as np
x = np.linspace(0, 30, 500)
y1 = np.exp(-0.1*x)*np.cos(x)
y2 = np.exp(-0.2*x)*np.sin(x)
plt.plot(x, y1, "b-", label="Partícula 1")
plt.plot(x, y2, "r-", label="Partícula 2")
plt.xlabel("Tiempo (s)")
plt.ylabel("Posición (mm)")
plt.title("Gráfica de posición")
plt.legend()
plt.show()
```
